# データの取得

```python
# ライブラリの読み込み
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from pathlib import Path
```

```python
# サンプルデータセットの読み込み
data = load_breast_cancer()
```

```python
# pandas Dataframe形式で説明変数Xを取得
df_X = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
# pandas Series形式で目的変数yを取得
s_y = pd.Series(data.target)
```

```python
# pandas Dataframe形式で目的変数と説明変数を結合したXyを取得
df_Xy = pd.concat([df_X, s_y], axis=1, join='inner')
```

```python
# colum最終列にある目的変数に名前がついていないので"target'という名前に変更
df_Xy = df_Xy.rename(columns={0:"target"})
```

# データセットの概要を確認

```python
# データセットの最初の10行くらいをみてみる
df_Xy.head(10)
```

```python
# 説明変数の種類、欠損値の有無、データ型を確認
df_Xy.info()
```

```python
# 欠損値の有無に着目したいときはisnull()関数を使う
df_Xy.isnull().any()
```

```python
# 説明変数ごとの平均・標準偏差、４分位数の確認
df_Xy.describe()
```

```python
# カテゴリー変数の場合は、uniqueな値を確認しておく。（unique関数はSeries形式に対して使う）
df_Xy['target'].unique()
```

```python
# カテゴリー変数の場合は、カテゴリごとの数を確認してみる
df_Xy['target'].value_counts()
```

```python
# 相関行列
corr_mat = df_Xy.corr()
corr_mat
```

# データの可視化

```python
'''
特徴量ごとのヒストグラムを描写する
特徴量が少ない場合は

sns.pairplot(data=df_Xy, hue='target', kind='reg').savefig('pairplot.png', facecolor='white')

といった形ですべてのヒストグラムと散布図の組み合わせを一括で確認することも出来る
'''


for i, col in enumerate(df_Xy.columns):
    # ヒストグラムの描写
    sns.histplot(df_Xy[df_Xy['target'] == 0][col], kde=True, color='salmon')
    sns.histplot(df_Xy[df_Xy['target'] == 1][col], kde=True, color='royalblue')
    labels = ['Target=0(kde)', 'Target=1(kde)', 'Target=0(Count)', 'Target=1(Count)']
    plt.legend(labels)
    folder = Path('output/hist')
    filename = col+'_hist.png'
    file = folder / filename
    plt.savefig(file)
    plt.figure()
    
```

```python
# 相関行列からヒートマップを作成
plt.figure(figsize=(25,25)) # 出力する画像サイズの調整
sns.heatmap(corr_mat,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )
folder = Path('output/heatmap')
filename = 'heatmap.png'
file = folder / filename
plt.savefig(file)
plt.figure()
```
